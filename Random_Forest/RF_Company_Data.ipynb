# Problem Statement:

A cloth manufacturing company is interested to know about the segment or attributes causes high sale. 
Approach - A Random Forest can be built with target variable Sales (we will first convert it in categorical variable) & all other variable will be independent in the analysis.  


### About the data: 


Letâ€™s consider a Company dataset with around 10 variables and 400 records. 
The attributes are as follows: 

- Sales -- Unit sales (in thousands) at each location
- Competitor Price -- Price charged by competitor at each location
- Income -- Community income level (in thousands of dollars)
- Advertising -- Local advertising budget for company at each location (in thousands of dollars)
- Population -- Population size in region (in thousands)
- Price -- Price company charges for car seats at each site
- Shelf Location at stores -- A factor with levels Bad, Good and Medium indicating the quality of the shelving location for the car seats at each site
- Age -- Average age of the local population
- Education -- Education level at each location
- Urban -- A factor with levels No and Yes to indicate whether the store is in an urban or rural location
- US -- A factor with levels No and Yes to indicate whether the store is in the US or not

### Independent Variables

```python
compX = comp1.iloc[:,1:]
compX
```

```python
# It will exclude original categorical columns and will include only included columns having dummy values. 

compX_encoded = pd.get_dummies(compX, columns = ['ShelveLoc', 'Urban', 'US'])
compX_encoded.head()
```

### Dependent Variable

```python
comp1.head()
```

```python
comp1.loc[comp1["Sales"] <= 10.00,"Sales1"]="Not High"
comp1.loc[comp1["Sales"] >= 10.01,"Sales1"]="High"

comp1.head()
```

```python
comp1['Sales1'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder() 
```

```python
compY = lb.fit_transform(comp1.Sales1)
```

```python
compY.shape
```

```python
compY
```

```python
comp1['Sales1'].value_counts()
```

```python
compX_encoded.shape
```

# Building Preliminary Model

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(compX_encoded, compY, test_size = 0.25, random_state = 25)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=40)
```

```python
rf.fit(x_train, y_train)
```

```python
rf.get_params()
```

```python
y_pred = rf.predict(x_test)
y_pred
```

### Model Evaluation

```python
rf.score(x_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```

### GridSearch Method

```python
from sklearn.model_selection import GridSearchCV


# Create the parameter grid based on the results of random search 
param_grid = {
    'bootstrap': [True],
    'max_depth': [20, 40, 50, 60],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [50, 100, 150, 200, 220]
}
```

```python
# Create a based model
rf = RandomForestClassifier()
```

```python
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2, refit = True)
```

```python
# Fit the grid search to the data
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
grid_pred = grid_search.predict(x_test)
grid_pred
```

#### Model Evaluation

```python
print(classification_report(y_test, grid_pred))
```

```python
print(confusion_matrix(y_test, grid_pred))
```
